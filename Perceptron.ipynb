```python
import matplotlib.pyplot as plt
import numpy as np
import random
plt.rcParams['font.size']=14
```

### Data

```python
# Inputs and Labels
x = [[0.72, 0.82,-1],
    [0.91,-0.69,-1],
    [0.03,0.93,-1],
    [0.12,0.25,-1],
    [0.96,0.47,-1],
    [0.8,-0.75,-1],
    [0.46,0.98,-1],
    [0.66,0.24,-1],
    [0.72,-0.15,-1],
    [0.35,0.01,-1],
    [-0.11,0.1,1],
    [0.31,-0.96,1],
    [0.0,-0.26,1],
    [-0.43,-0.65,1],
    [0.57,-0.97,1],
    [-0.72,-0.64,1],
    [-0.25,-0.43,1],
    [-0.12,-0.9,1],
    [-0.58,0.62,1],
    [-0.77,-0.76,1]]

x = np.array(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = x[:, 2])
plt.show()
```

### Perceptron Training

_The Weights are initialized as zeros, the Threshold and Bias are respectively set to 0 and 1_

```python
# Weights (Synapses)
w = [0, 0]
# Bias
bias = 1
# Threshold
threshold = 0
# Learning rate - [0, 1]
learning_rate = 0.1
# Maximum number allowed of iterations
max_iterations = 5
```

```python
for num_iter in range(1, max_iterations+1):
#   Getting predictions
    num_correct = 0
    for ind in range(len(x)):
        z = x[ind][0]*w[0] + x[ind][1]*w[1] + bias
        
        if z >= threshold:
            y=1
        else:
            y=-1
            
        if y == x[ind][2]:
            num_correct += 1
        # Update the Weights    
        else:
            w[0] = w[0] + learning_rate * (x[ind][2]- y) * x[ind][0] 
            w[1] = w[1] + learning_rate * (x[ind][2]- y) * x[ind][1]
            bias = bias + learning_rate * (x[ind][2]- y)
            
    print('iter:', num_iter, 'num_correct: ', num_correct, 'W1: ', round(w[0], 2), 'W2: ', round(w[1], 2), 'Bias: ', round(bias,2))            
```
